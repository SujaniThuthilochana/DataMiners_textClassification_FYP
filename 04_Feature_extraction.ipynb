#Sentiment using TextBlob

```python
from textblob import TextBlob

# Define a function that takes a piece of text as input and returns the sentiment label
def get_sentiment_label(text):
  blob = TextBlob(text)
  sentiment = blob.sentiment
  polarity = sentiment.polarity
  if polarity > 0:
    return "positive"
  elif polarity < 0:
    return "negative"
  else:
    return "neutral"

# Apply the function to the 'text' column of the DataFrame and store the results in a new 'sentiment' column
new_f['sentiment'] = new_f['Preprocessed tweet'].apply(get_sentiment_label)
```

#Word Cloud for All lemmas

```python
text_list = new_f['Preprocessed tweet']


# Next, concatenate all the strings into a single string
text = ' '.join(text_list)

# Next, create a WordCloud object
from wordcloud import WordCloud
wordcloud = WordCloud().generate(text)

# Finally, display the wordcloud using Matplotlib
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
